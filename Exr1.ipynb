```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
Bhing = cv2.imread('1.jpg')
RGB = cv2.cvtColor(Bhing, cv2.COLOR_BGR2RGB)

plt.imshow(RGB)
```

```python
backBhing = cv2.imread('Sky.jfif')
backRGB = cv2.cvtColor(backBhing, cv2.COLOR_BGR2RGB)
plt.imshow(backRGB)
```

```python
rows, cols, _ = backBhing.shape
Bhing = cv2.resize(Bhing, (cols, rows))

RGB = cv2.cvtColor(Bhing, cv2.COLOR_BGR2RGB)
HSV = cv2.cvtColor(Bhing, cv2.COLOR_BGR2HSV)

plt.imshow(RGB)
```

```python
H, S, V = cv2.split(HSV)

plt.imshow(H, cmap='gray')
plt.title('H')
plt.show()

plt.imshow(S, cmap='gray')
plt.title('S')
plt.show()

plt.imshow(V, cmap='gray')
plt.title('V')
plt.show()
```

```python
for i in range(rows):
    for j in range(cols):
        if 25 < H[i, j] < 75 and S[i, j] > 60 and V[i, j] > 80:
            RGB[i, j] = backRGB[i, j]
            
plt.imshow(RGB)
```
